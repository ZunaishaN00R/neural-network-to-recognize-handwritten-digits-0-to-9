```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
```

Load the MNIST dataset

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

Normalize pixel values to be between 0 and 1

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

Build the model

```python
model = models.Sequential()
```

Flatten layer to convert 28x28 images to a flat vector

```python
model.add(layers.Flatten(input_shape=(28, 28)))
```

Experiment with different architectures and activation functions

```python
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

Compile the model

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Train the model

```python
history = model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```

Evaluate the model on the testing set

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test Accuracy:", test_acc)
```

Visualize the model's predictions on random test samples

```python
predictions = model.predict(test_images)
```

Choose random test samples

```python
num_samples_to_display = 5
random_indices = np.random.choice(len(test_images), num_samples_to_display, replace=False)
```

Display the images and their predictions

```python
for i, index in enumerate(random_indices):
    plt.subplot(1, num_samples_to_display, i + 1)
    plt.imshow(test_images[index], cmap='gray')
    plt.title(f"Predicted: {np.argmax(predictions[index])}")
    plt.axis('off')

plt.show()
```
